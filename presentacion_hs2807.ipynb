```python
enero = pd.DataFrame({"Nombre":["Juan","Romina"],"Deuda":[-10000,0],"Score":[300,800]})
febrero = pd.DataFrame({"Nombre":["Juan","Romina"],"Deuda":[-15000,-500],"Score":[400,800]})
```

```python
enero
```

```python
febrero
```

```python
porcentaje = []

for n in enero.index:
    deuda_uno = enero.loc[n,"Deuda"]
    deuda_dos = febrero.loc[n,"Deuda"]
    if abs(deuda_uno) < abs(deuda_dos):
        if abs(deuda_uno) != 0:
            r = abs(deuda_dos) / abs(deuda_uno)
            r = r - 1
            r = r * 100
        else:
            r = 100
        
    porcentaje.append(r)
    

porcentaje
        
```

```python
febrero["Porcentaje"] = porcentaje
febrero
```
